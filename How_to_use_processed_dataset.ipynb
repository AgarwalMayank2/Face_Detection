Loading dataset and making dataframe from github url (raw CSV file)

```python
import pandas as pd
```

# Loading filtered_CNN_features_dataset

```python
url_filtered_CNN_features_dataset = 'https://raw.githubusercontent.com/AgarwalMayank2/Face_Detection/refs/heads/main/processed_dataset/filtered_CNN_features_dataset.csv' #for CNN limited
df = pd.read_csv(url_filtered_CNN_features_dataset) # reading url for extracted CNN_features_dataset_limited.csv
```

```python
df.head(30)
```

Splitting dataframe into train and test in 4:1 ratio

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Separate features and labels
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Encode labels (alphabetically)
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
# encoding is benificial as working on numbers is lot easier than working on string

# Ensure stratified split (16 training, 4 testing per class)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=1/5, random_state=42, stratify=y_encoded)

# Normalize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(f"Dataset size: {df.shape}")
print(f"Training size: {X_train.shape}, Testing size: {X_test.shape}")
```

how to decode encoded label :

```python
# Decode labels
y_decoded = label_encoder.inverse_transform([0, 1, 157, 3])
print(y_decoded)
```
